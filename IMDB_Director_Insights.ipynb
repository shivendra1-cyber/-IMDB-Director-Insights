```python
# a new column is_risky 
# is_risky --> whether a movie was risk for the director (True/False)
#risky movie is define as movie with average budget higher than mean of director's budget 
def cal_risky(sub_df):
   sub_df["risky"]= sub_df["budget"]-sub_df["revenue"].mean() >=0
   return sub_df
data_risky=data.groupby('director_name').apply(cal_risky)

```

```python
agg_dict = {'year':['min','max'],'title':'count'}
data_agg=data.groupby('director_name')[['year','title']].aggregate(agg_dict)
data_agg
```

```python
# [year_min,year_max,title_count]
data_agg=['_'.join(tup) for tup in data_agg]
```

```python
data_agg=data.groupby('director_name')[['year','title']].aggregate(
    year_max=('year','max'),year_min=('year','min'), no_movies=('title','count'))
data_agg
```

```python
data_agg['yrs_active'] = data_agg['year_max']- data_agg['year_min']
data_agg
```

```python
data_agg
```

```python
data_agg['productivity']= data_agg['no_movies'] / data_agg["yrs_active"]
data_agg
```
